```python
%matplotlib inline
import sympy as sym
from sympy import Symbol,pprint
import sympy.plotting as syp
```

```python
sigma=Symbol('sigma')
x=Symbol('x')
mu=Symbol('mu')
```

```python
part_1=1/(sym.sqrt(2*sym.pi*sigma**2))
part_2=sym.exp(-1*((x-mu)**2)/(2*sigma**2))
my_gauss_function=part_1*part_2
```

```python
pprint(my_gauss_function)
```

```python
syp.plot(my_gauss_function.subs({mu:10,sigma:30}),(x,-1000,1000),title='gaus distribution')
```

```python
x_values=[]
y_values=[]
for value in range(-50,50):
    y_values.append(my_gauss_function.subs({mu:0,sigma:10,x:value}).evalf())
    x_values.append(value)
import matplotlib.pyplot as plt
plt.plot(x_values,y_values)
```
